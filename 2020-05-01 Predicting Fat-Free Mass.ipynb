# Relating Fat Mass to Fat-Free Mass: Did This Study Overfit their Model?

In the Diana Thomas-led study "Research New fat free mass - fat mass model for use in
physiological energy balance equations", the scientists claim that the long-used Forbes equation that relates Fat Mass to Fat-Free Mass is outdated.  The study can be found here:

https://nutritionandmetabolism.biomedcentral.com/articles/10.1186/1743-7075-7-39

They developed a class of 4th-order polynomial equations that they claim is more accurate than the forbes equation.  However, a 4th-order polynomial equation is very complicated, and it is possible that these equations are overfit to the data.

# Methodology

Since the study was published in 2010, any data past that point is fair game to test the model on.  I will be using 2017-2018 Laboratory Demographic Data from the NHANES website to test the model, which can be found here:

https://wwwn.cdc.gov/nchs/nhanes/continuousnhanes/default.aspx?BeginYear=2017

I imported the data, engineered the necessary features, and ran the data through the four 4th-order polynomial models that calculate fat-free mass.  Then, I compared the predictions of the model to the actual fat-free masses of the participants in the 2017-2018 study, testing the accuracy of the model on new data to see if the models were overfit.

Finally, I created and tested a new model relating Fat Mass to Fat-Free Mass using common machine learning guidelines to avoid overfitting and more successfully predict Fat-Free Mass.

# Model Repoductions

First, I must make sure I can succesffully reproduce the models in the study, before I try testing them for overfitness.

The Thomas models are separated by gender and race: One for African American females, African American males, non-African American females, and non-African American males.

## African American Female Fat-Free Mass Model

```python
def african_american_female_ffm_model(df):
    f = df['fatmass']
    a = df['age']
    h = df['height'] 
    fat_free_mass = (-69 + 2.5*f - 0.04*a + 0.3*h - 0.002*f*a - 0.01*f*h - 0.047*f**2 
                     + 0.00003*f**2*a + 0.0000004*f**4 + 0.0002*f**3 + 0.0003*f**2*h - 0.000002*f**3*h)
    return fat_free_mass
```

### Explanation of Input and Output of a Function

A function receives input features, called **parameters**, that will be used inside the function.  In this case, the parameter is the DataFrame **df**.  A function can also return a value, which will become the value of the function call later in your code.  In this case, the function returns **fat_free_mass**.

## African American Male Fat-Free Mass Model

```python
def african_american_male_ffm_model(df):
    f = df['fatmass']
    a = df['age']
    h = df['height'] 
    fat_free_mass = (-69 + 3.6*f - 0.04*a + 0.7*h - 0.002*f*a - 0.01*f*h + 0.00003*f**2*a 
                     - 0.07*f**2 + 0.007*f**3 - 0.000002*f**4 + 0.0003*f**2*h - 0.000002*f*h)
    return fat_free_mass
```

## Non-African American Female Fat-Free Mass Model

```python
def non_african_american_female_ffm_model(df):
    f = df['fatmass']
    a = df['age']
    h = df['height'] 
    fat_free_mass = (-72.1 + 2.5*f - 0.04*a + 0.7*h - 0.002*a - 0.01*f*h - 0.04*f**2
                     + 0.00003*f**2*a + 0.0000004*f**4 + 0.0002*f**3 + 0.0003*f**2*h - 0.000002*f**3*h)
    return fat_free_mass
```

## Non-African American Male Fat-Free Mass Model

```python
def non_african_american_male_ffm_model(df):
    f = df['fatmass']
    a = df['age']
    h = df['height'] 
    fat_free_mass = (-71.7 + 3.6*f - 0.04*a + 0.7*h - 0.002*f*a - 0.01*f*h + 0.00003*f**2*a
                     - 0.07*f**2 + 0.0006*f**3 - 0.000002*f**4 + 0.0003*f**2*h - 0.000002*f**3*h)
    return fat_free_mass
```

# Testing for Overfitness

In a usual supervised regression analysis, where we have all of the input features and target variable, I would split the data into train and test sets, and develop a model using the train dataset that predicts the target variable well in the test dataset.  However, the models have already been developed in the study using pre-2010 data, so I just need to test them on the 2017-2018 dataset.

## Scoring Metric: Mean Squared Error (MSE)

```python
from sklearn.metrics import mean_squared_error
```

SKLearn is a Python library that provides 'simple and efficient tools for predictive data analysis'.  The full documentation can be found here:

https://scikit-learn.org/stable/

Mean Squared Error calculates the average squared difference between your predicted target values and actual target values.  A low error rate would indicate that the models are not overfit, and they still predict 2017-2018's data well.  A high error rate would be a red flag that the models were overfit, since they no longer accurately predict Fat-Free Mass.

## Finding MSE For Each Model

### MSE AA Females

```python
y_pred = african_american_female_ffm_model(df.loc[(df['gender']=='female')&(df['race']=='AA')])
y_true = df['fatfreemass'].loc[(df['gender']=='female')&(df['race']=='AA')]
MSE_AA_F = mean_squared_error(y_true,y_pred)
```

### MSE AA Males

```python
y_pred = african_american_male_ffm_model(df.loc[(df['gender']=='male')&(df['race']=='AA')])
y_true = df['fatfreemass'].loc[(df['gender']=='male')&(df['race']=='AA')]
MSE_AA_M = mean_squared_error(y_true,y_pred)
```

### MSE NAA Females

```python
y_pred = non_african_american_female_ffm_model(df.loc[(df['gender']=='female')&(df['race']=='NAA')])
y_true = df['fatfreemass'].loc[(df['gender']=='female')&(df['race']=='NAA')]
MSE_NAA_F = mean_squared_error(y_true,y_pred)
```

### MSE NAA Males

```python
y_pred = non_african_american_male_ffm_model(df.loc[(df['gender']=='male')&(df['race']=='NAA')])
y_true = df['fatfreemass'].loc[(df['gender']=='male')&(df['race']=='NAA')]
MSE_NAA_M = mean_squared_error(y_true,y_pred)
```

## Displaying MSE Results

```python
mse_dict = {
                'African American Females':np.round(MSE_AA_F,3),
                'African American Males':np.round(MSE_AA_M,3),
                'Non-African American Females':np.round(MSE_NAA_F,3),
                'Non-African American Males':np.round(MSE_NAA_M,3),
                'Average':np.round(np.mean([MSE_AA_F,MSE_AA_M,MSE_NAA_F,MSE_NAA_M]),3)
           }
```

# Creating a New Model that is not Overfit

Using Linear Regression, I will attempt to create a new model that may not as accurately predict Fat-Free Mass in the data used in the study, but will more accurately predict Fat-Free Mass in the 2017-2018 data and future NHANES data collection.

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

I will use the SKLearn Library again to run a regression on the data.  While there are other ways this can be done, such as using an OLS model, SKLearn is what I am most familiar with.

## Selecting Data Features

```python
X = df.drop(['fatfreemass','ID'],axis=1)
y = df['fatfreemass']
```

X will be the same input features used in the Thomas models: Fat Mass, Age, Height, Gender, and Race, while y will be the same target feature in the Thomas models: Fat Free Mass.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(f"X train shape: {X_train.shape}")
print(f"X test  shape: {X_test.shape}")
```

I split the data 80/20 into train and test sets.

```python
numerical_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object', 'bool']).columns
print(f'Numerical features are: {numerical_features}')
print(f'Categorical features are: {categorical_features}')
```

## Scaling Numerical Features

Before running numerical features through a Linear Regression, you can set them all to the same scale, so that larger numbers will not have a larger weight on the model.

```python
numerical_features = ['fatmass','age','height']
numerical_pipeline = Pipeline([('scaler',StandardScaler())])
```

## One Hot Encoding Categorical Features

In order to run categorical features through a Linear Regression, you must first convert them to numeric values.  This can be done using One Hot Encoding (OHE).  OHE is a method that splits categorical features into multiple columns, where their value will be binary, based on if the test subject has that value or not.

```python
categorical_features = ['gender','race']
categorical_pipeline = Pipeline([('OHE',OneHotEncoder())])
```

## Transforming Features

Using SKLearn Pipelines, I can transform the features of the model in order so that I can successfully run the regression.

```python
data_pipeline = ColumnTransformer( transformers= [
        ("numerical_pipe", numerical_pipeline, numerical_features),
        ("categorical_pipe", categorical_pipeline, categorical_features),
    ], 
        n_jobs=-1 #Minimizes runtime by using all cores of machine
    )
X_train_transformed = data_pipeline.fit_transform(X_train)
column_names = numerical_features  + \
               list(data_pipeline.transformers_[1][1].named_steps["OHE"].get_feature_names(categorical_features))

display(pd.DataFrame(X_train_transformed,  columns=column_names).head())
```

As you can see, the numerical features have been scaled and the categorical features have been OHE'd.

## Running the Linear Regression

Now that the features are ready for modeling, I can actually run the regression and analyze the MSE of the new model.

```python
clf_pipe = Pipeline([                       #Preprocesses the data and runs Linear Regression on it
            ('data_pipe',data_pipeline),
            ('LR',LinearRegression())
        ])
clf_pipe.fit(X_train, y_train)              #Trains the model
y_train_preds = clf_pipe.predict(X_train)   #Predicts Train Set's Fat-Free Masses based on trained model
y_test_preds = clf_pipe.predict(X_test)     #Predicts Test Set's Fat-Free Masses based on trained model
```

```python
MSE_train = mean_squared_error(y_train,y_train_preds)
MSE_test = mean_squared_error(y_test,y_test_preds)
```

```python
mse_dict = {
                'Train':np.round(MSE_train,3),
                'Test':np.round(MSE_test,3)
           }
```

```python
df_mse = pd.DataFrame(mse_dict,index=['MSE'])
df_mse
```

Because the MSE of the Test set's predictions of Fat-Free Mass is very close to the MSE of the Train set's predictions, the model is not overfit.  That means this model is already more predictive than the Thomas models.

# Retrieving the New Model

Now that I have created a model for predicting Fat-Free Mass that is not overfit, I would like to know the equation of the model.

```python
model = clf_pipe.named_steps['LR']
print(f'Columns are {column_names}')
print(f'Intercept is {np.round(model.intercept_,4)}')
print(f'Coefficients are {np.round(model.coef_,4)}')
```

So, using the Intercept and the coefficients, the equation for the new model is:

Female, Male, AfricanAmerican, and NonAfricanAmerican equal 0 or 1 depending on the demographics of the subject, instead of having seperate models for each combination of race and gender.
